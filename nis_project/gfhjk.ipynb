```python
import re
import os

# создает таблицу
with open('sentences.csv', 'w', encoding='utf-8') as file: 
    row = '%s\t%s\t%s\t%s\t%s\t%s\t%s\n'
    file.write(row % ('Path', 'Full sentence', 'Tagged mistake', 'Tag', 'Corrected mistake', 'Group', 'Correct/incorrect'))
path = './data/' # прописывает путь
# обходит дерево
for root, dirs, files in os.walk(path):
    for file in files:
        counter = 0 # счетчик
        if ".ann" in file: # для файлов с нужным расширением
            filename = str(file)
         #   print(root, dirs, file)
            final_ann = os.path.join(root, file) # полный путь к файлу
            with open(final_ann, encoding="utf-8") as file:
                annot_txt = file.read()
          #  print(annot_txt)
            clean_text = annot_txt.splitlines() # список строк
            for i in range(len(clean_text)-1): # обход строк в списке
              #  print(line)
                find_tag = re.search(r'T.*?Tense_choice.*?', clean_text[i]) # для строки с нужным тегом
                if find_tag is not None:
                    tag_id = re.search(r'(T\d*?)\s', clean_text[i])
                    identif = tag_id.group(1)
                    tag_id_path = os.path.join(final_ann[7:-4], identif)
                    print(tag_id_path)
                    tag_mistake = re.search(r'T\d.*\d\s(.*)', clean_text[i], flags=re.DOTALL) # ошибка
                    mistake = tag_mistake.group(1)
                    tag_correct = re.search(r'#.*T\d*\s(.*)', clean_text[i+1], flags=re.DOTALL)
                    if tag_correct is None:
                        correct = ''
                    else:
                        correct = tag_correct.group(1)
                    tag_place = re.search(r'\s\d.*\d\s', clean_text[i], flags=re.DOTALL)
                    position = tag_place.group()
                    print(position)
                    position_list = position.split()
                    if len(position_list) == 2:
                        pos1 = position_list[0]
                        pos2 = position_list[1]
                        position2 = int(pos2) + 300
#                     pos1,pos2 = position.split()
                        text_file  = filename[0:-3] + 'txt'
                        final_text = os.path.join(root, text_file)
                        with open(final_text, encoding="utf-8") as f:
                            text = f.read()
                        text = text.replace('\n','')
                        text_lines = text.split('.')
                        for line in text_lines:
                         #   counter += len(line)
                         #   counter += 1
                         #   print(counter)
                         #   if counter >= int(pos1) and counter <= int(position2):
                         #       sentence_with_tag = line
                         #       print('\n\n', counter, line)
                            if text.find(line)<=int(pos1) and text.find(line)+len(line)>int(pos2):
                                sentence_with_tag = line.replace('\t', '')
                    # пополенение таблицы
                        with open('sentences.csv', 'a', encoding='utf-8') as f:
                            row = '%s\t%s\t%s\tTense_choice\t%s\t%s\t%s\n'
                            f.write(row % (final_ann, sentence_with_tag, mistake, correct, ' ', ' '))
```
